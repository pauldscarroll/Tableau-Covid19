```python
def converter(x):
    if type(x) == float:
        return int(x)
    else:
        pass

```

```python
train.ConfirmedCases = train.ConfirmedCases.apply(converter)
```

```python
train.Fatalities = train.Fatalities.apply(converter)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content)
```

```python
countries = soup.find_all("table")[0]
df = pd.read_html(str(countries))[0]
```

```python
df.head(30)
```

```python
df.rename(columns=rename_map, inplace=True)
df.head(2)
```

```python
country = df[['Country','Population']]
country.head(2)
```

```python
country = country[(country['Population'] > 5000000)]
country.head()
```

```python
train.Country.unique()
```

```python
country.Country.unique()
```

```python
country_dict = {'United States':'US',
            'Czechia':'Czech Republic (Czechia)'}
country['Country'] = country.Country.replace(country_dict)
```

```python
country.head()
```

```python
kgworked = pd.merge(train, country,on='Country')[['Country', 'Population','Fatalities','ConfirmedCases','Date']]
kgworked.head()
```

```python
kgworked['Date'] = pd.to_datetime(kgworked['Date'], format = '%Y-%m-%d')
#test['Date'] = pd.to_datetime(test['Date'], format = '%Y-%m-%d')

kgworked.set_index('Date',inplace=True,drop=True)
kgworked.head(2)
```

```python
kgworked['Confirmedper1000']= kgworked.ConfirmedCases/ (kgworked.Population/1000)
```

```python
kgworked
```

```python
kgworked.to_csv('trained.csv')
```
